**Estudo de Caso 1:**
Como permitir o uso dos operadores aritméticos e de comparação do Python (+, -, *, /, >, <, ==, !=, <=, >=) entre objetos fração e imprimir os objetos corretamente utilizando print(objeto).

```python
class Fracao:

  __numerador = None
  __denominador = None

  def __init__(self, numerador, denominador):
    if denominador == 0:
      raise ZeroDivisionError('Erro de divisão por zero!!')
    else:
      self.numerador = numerador
      self.denominador = denominador

  @property
  def numerador(self):
    return self.__numerador

  @property
  def denominador(self):
    return self.__denominador

  @numerador.setter
  def numerador(self, numerador):
    self.__numerador = numerador

  @denominador.setter
  def denominador(self, denominador):
    if denominador == 0:
      raise ZeroDivisionError('Erro de divisão por zero!!')
    else:
      self.__denominador = denominador

  def __str__(self):
        return f'{self.numerador}/{self.denominador}'

  def __add__(self, fracao):
    if fracao.denominador == self.denominador:
      return Fracao(fracao.numerador + self.numerador,fracao.denominador )
    else:
      return Fracao(fracao.numerador*self.denominador+self.numerador*fracao.denominador,fracao.denominador*self.denominador )

  def __sub__(self, fracao):
    if fracao.denominador == self.denominador:
      return Fracao(fracao.numerador - self.numerador,fracao.denominador )
    else:
      return Fracao(fracao.numerador*self.denominador-self.numerador*fracao.denominador,fracao.denominador*self.denominador )

  def __mul__(self, fracao):
    return Fracao(fracao.numerador*self.numerador, fracao.denominador*self.denominador)

  def __truediv__(self, fracao):
    return Fracao(self.numerador*fracao.denominador, self.denominador*fracao.numerador)

  def __float__(self):
    return int(self.numerador)/int(self.denominador)

  def __lt__(self, fracao):
    """ Menor que """
    return float(self) < float(fracao)

  def __le__(self, fracao):
    """ Menor ou igual """
    return float(self) <= float(fracao)

  def __eq__(self, fracao):
    return float(self) == float(fracao)

  def __ne__(self, fracao):
    """ Diferente """
    return float(self) != float(fracao)

  def __ge__(self, fracao):
    """ Maior ou igual """
    return float(self) >= float(fracao)

  def __gt__(self, fracao):
    """ Maior que """
    return float(self) > float(fracao)

fracao1 = Fracao(15,8)
fracao2 = Fracao(3,1)
```

```python
print(f"Operação: {fracao1} + {fracao2} = {fracao1+fracao2}")
print(f"Operação: {fracao1} - {fracao2} = {fracao1-fracao2}")
print(f"Operação: {fracao1} * {fracao2} = {fracao1*fracao2}")
print(f"Operação: {fracao1} / {fracao2} = {fracao1/fracao2}")
print(f"Comparação: {fracao1} < {fracao2} => Resultado: {fracao1<fracao2}")
print(f"Comparação: {fracao1} <= {fracao2} => Resultado: {fracao1<=fracao2}")
print(f"Comparação: {fracao1} == {fracao2} => Resultado: {fracao1==fracao2}")
print(f"Comparação: {fracao1} != {fracao2} => Resultado: {fracao1!=fracao2}")
print(f"Comparação: {fracao1} >= {fracao2} => Resultado: {fracao1>=fracao2}")
print(f"Comparação: {fracao1} > {fracao2} => Resultado: {fracao1>fracao2}")
```
