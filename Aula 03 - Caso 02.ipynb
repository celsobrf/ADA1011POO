```python
## CRIANDO TODOS OS ARQUIVOS NO AMBIENTE UTILIZANDO AS BIBLIOTECAS SYS E OS

import sys
import os

sys.path.append('/content')
os.makedirs('/content/matematica')

arquivo_fracoes = open("matematica/Fracoes.py", "w")
arquivo_geometricas = open("matematica/Geometricas.py", "w")
arquivo_pacote = open("matematica/__init__.py", "w")

arquivo_fracoes.write(
"""class Fracao:

  def __init__(self, numerador, denominador):
    self.numerador = numerador
    self.denominador = denominador

  def __str__(self):
        return f'{self.numerador}/{self.denominador}'

  def soma(self, fracao):
    if fracao.denominador == self.denominador:
      return Fracao(fracao.numerador + self.numerador,fracao.denominador )
    else:
      return Fracao(fracao.numerador*self.denominador+self.numerador*fracao.denominador,fracao.denominador*self.denominador )

  def __add__(self, fracao):
    return self.soma(fracao)

  def subtracao(self, fracao):
    if fracao.denominador == self.denominador:
      return Fracao(fracao.numerador - self.numerador,fracao.denominador )
    else:
      return Fracao(fracao.numerador*self.denominador-self.numerador*fracao.denominador,fracao.denominador*self.denominador )

  def __sub__(self, fracao):
    return self.subtracao(fracao)

  def multiplicacao(self, fracao):
    return Fracao(fracao.numerador*self.numerador, fracao.denominador*self.denominador)

  def __mul__(self, fracao):
    return self.multiplicacao(fracao)

  def divisao(self, fracao):
    return Fracao(self.numerador*fracao.denominador, self.denominador*fracao.numerador)

  def __truediv__(self, fracao):
    return self.divisao(fracao)""")

arquivo_geometricas.write(
"""class Retangulo:

  def __init__(self, x, y):
    self.x = x
    self.y = y

  def area(self):
    return self.x*self.y

class Quadrado:

  def __init__(self, lado):
    self.lado = lado

  def area(self):
    return self.lado**2

class Triangulo:

  def __init__(self, base, altura):
    self.base = base
    self.altura = altura

  def area(self):
    return self.base*self.altura/2

class Circulo:

  def __init__(self, raio):
    self.raio = raio

  def area(self):
    return 3.14*self.raio**2""")

arquivo_pacote.write(
    """from .Fracoes import Fracao
from .Geometricas import Retangulo
from .Geometricas import Quadrado
from .Geometricas import Triangulo
from .Geometricas import Circulo""")


arquivo_fracoes.close()
arquivo_geometricas.close()
arquivo_pacote.close()
```
